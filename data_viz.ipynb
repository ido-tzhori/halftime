```python
import plotly.express as px
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('f_eng_df.csv', index_col = [0]).reset_index(drop = True).fillna(0)
```

```python
df.columns
```

```python
pca = PCA(n_components=2)
components = pca.fit_transform(df.iloc[:,:-1])

fig = px.scatter(components, x=0, y=1, color=df['target'])
fig.show()
```

```python
f, ax = plt.subplots(figsize=(10,10))
sns.heatmap(df.iloc[:,:20].corr(), annot=True, linewidth=".5", cmap="RdPu", fmt=".2f", ax = ax)
plt.title("Correlation Map",fontsize=20)
plt.show()
```
