```python
from pycaret.classification import *
import pandas as pd
import pickle
```

```python
df = pd.read_csv('og_half.csv', index_col = [0]).drop(['fh_winner','Under Pressure','Yellow Cards','Fouls Wons','Red Cards'], axis = 1)
```

```python
df
```

```python
grid = setup(data=df, target='target', feature_selection = True, remove_multicollinearity = True)
```

```python
best = compare_models()
evaluate_model(best)
```

```python
best
```

```python
lr = create_model('lr')
evaluate_model(lr)
```

```python
tuned = tune_model(lr,n_iter = 30)
evaluate_model(tuned)
```

```python
preds = predict_model(tuned, raw_score = True)
```

```python
tuned.predict_proba([np.zeros(9)])
```

```python
preds.groupby('Goals').count()
```

```python
finalize_model(tuned)
```

```python
save_model(tuned, 'lr_model')
```
